```python
x1 = data.loc[:,['Age', 'Total Salary']].values
from sklearn.cluster import KMeans
wcss= []
for k in range(1,11):
  kmeans= KMeans(n_clusters=k, init='k-means++')
  kmeans.fit(x1)
  wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color='red',marker='8')
plt.xlabel('K Value')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
label = kmeans.fit_predict(x1)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x1[:,0],x1[:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.title('cluster of customers')
plt.xlabel('Age')
plt.ylabel('Total salary')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(x1)
print(label)
```

```python
clusters=kmeans.fit_predict(x5)
df['label']=clusters
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(15,10))
ax=fig.add_subplot(111, projection= '3d')
ax.scatter(df.Age[df.label==0], df['Total Salary'][df.label==0], df['Price'][df.label==0], c='blue', s=60)
ax.scatter(df.Age[df.label==1], df['Total Salary'][df.label==1], df['Price'][df.label==1], c='yellow', s=60)
ax.scatter(df.Age[df.label==2], df['Total Salary'][df.label==2], df['Price'][df.label==2], c='purple', s=60)
ax.view_init(30,185)

plt.xlabel('Age')
plt.ylabel('Total Salary')
ax.set_zlabel('Price')
plt.show()
```
