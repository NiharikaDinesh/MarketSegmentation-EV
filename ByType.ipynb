```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title('States and Two Wheelers(L1 & L2)')
pic=sns.barplot(x='State', y='Two Wheelers (Category L1 & L2 as per Central Motor Vehicles Rules', data=df,palette='ocean')
pic.set_xticklabels(pic.get_xticklabels(),rotation=90,ha='right')
```
